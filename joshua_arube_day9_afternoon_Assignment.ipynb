```python
"""
SERIES
1.	Write a Pandas program to compare the elements of two Pandas series. 
Sample Series [4,65,436,3,9], [7,0,3,897,9]
"""
import numpy as np
import pandas as pd

s1 = pd.Series([4, 65, 436, 3, 9])
s2 = pd.Series([7, 0, 3, 897, 9])

comparison = s1 == s2
print(comparison)
```

```python
"""
2.	Write a Pandas program to add, subtract, multiply and divide two Pandas series. 
Sample Series: [2,4,6,8,14], [1,3,5,7,9]

"""
series1 = pd.Series([2, 4, 6, 8, 14])
series2 = pd.Series([1, 3, 5, 7, 9])

addition = series1 + series2
subtraction = series1 - series2
multiplication = series1 * series2
division = series1 / series2

print("Addition:")
print(addition)

print("Subtraction:")
print(subtraction)

print("Multiplication:")
print(multiplication)

print("Division:")
print(division)
```

```python
"""
3.	Write a Pandas program to convert a dictionary to a Pandas series. 
Sample dictionary: dictionary1 = {‘Josh’: 24, ‘Sam’: 36, ‘Peace’: 19, ‘Charles’: 65, ‘Tom’: 44}

"""

dict = {'Josh': 24, 'Sam': 36, 'Peace': 19, 'Charles': 65, 'Tom': 44}

series = pd.Series(dict)
print(series)

```

```python
"""
4.	Write a Pandas program to convert a given series to an array.
Sample series: [‘Love’, 800, ‘Joy’, 789.9, ‘Peace’, True]

"""
series = pd.Series(['Love', 800, 'Joy', 789.9, 'Peace', True])

array = np.array(series)
print(array)
```

```python
"""
5.	Write a Pandas program to display the most frequent value in the given series and replace everything else as ‘Other’ in the series. (Use the ‘HomeTeamGoals’ column in the dataset provided)

"""

# Read the dataset into a Pandas DataFrame
afcon = pd.read_csv("AfricaCupofNationsMatches.csv")

# Extract the 'HomeTeamGoals' column
home_goals = afcon['HomeTeamGoals']

# Find the most frequent value
most_frequent_value = home_goals.mode()[0]

# Replace all other values with 'Other'
home_goals = home_goals.apply(lambda x: 'Other' if x != most_frequent_value else x)

# Print the updated Series
print(home_goals)
```

```python
"""
18.	Write a Pandas program to iterate over rows in your DataFrame.
"""
# Iterate over rows in the DataFrame
for index, row in df.iterrows():
    # Access values of each column for the current row
    year = row['Year']
    date = row['Date']
    home_team = row['HomeTeam']
    away_team = row['AwayTeam']
    home_goals = row['HomeTeamGoals']
    away_goals = row['AwayTeamGoals']

    
    print(f"Row {index}: {home_team} vs {away_team}, {home_goals}-{away_goals}, {date}, {year}")
```
